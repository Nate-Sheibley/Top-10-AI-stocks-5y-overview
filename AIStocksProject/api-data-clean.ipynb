```python
# Dependencies and setup
import pandas as pd
import os
```

```python
# kaggle csv path
kaggle_data = "Resources/kaggle_clean.csv"

# Read the Top 10 AI data
kaggle_df = pd.read_csv(kaggle_data)

# Display head of data to preview
kaggle_df.head()
```

```python
# API paths
DOW_data = "Resources/DOW"
INDU_data = "Resources/INDU"

# Read the Top 10 AI data
dow_df = pd.read_csv(DOW_data)
indu_df = pd.read_csv(INDU_data)

# Display head of data to preview
dow_df.head()
```

```python
#Rename the date column 
dow_df = dow_df.rename(columns={'Unnamed: 0': 'date'})
indu_df = indu_df.rename(columns={'Unnamed: 0': 'date'})
dow_df.head()
```

```python
# Define start date and end date of focus
start_date = "2022-06-01"
end_date = "2024-05-31"
```

```python
# Limit the data to the timeframe we want
DOW_cleaned = dow_df[(dow_df['date'] >= start_date) & (dow_df['date'] <= end_date)]
INDU_cleaned = indu_df[(indu_df['date'] >= start_date) & (indu_df['date'] <= end_date)]
```

```python
#define path to Resources folder
resources_path = "Resources"

#DOW
file_path1 = os.path.join(resources_path, 'dow_clean.csv')
DOW_cleaned.to_csv(file_path1, index=False, header=True)

#INDU
file_path2 = os.path.join(resources_path, 'indu_clean.csv')
INDU_cleaned.to_csv(file_path2, index=False, header=True)
```
